```python
import pickle
import numpy as np
import pyliftover
import tqdm
lo = pyliftover.LiftOver('hg19', 'hg38')    
```

```python
with open('snpNamesUKBBhapmap.pkl', 'rb') as f:
    commonList = pickle.load(f)
```

```python
newCoordList = []
oldCoordList = []
for chrIndex in range(1,23):
    nealePath = "/home/ozdemiro/LAB/learningData/HapmapProject/NealeGWAS/t2d_GWAS_chr" + str(chrIndex)
    with open(nealePath) as file:
        nealeData = file.read()
        nealeData = nealeData.split("\n")
        nealeData = nealeData[1:-1]
    nealeDict = {}
    nealeDictLetters = {}
    for x in range(len(nealeData)):
        dummyList = nealeData[x].split(" ")
        nealeDict[dummyList[0]]=dummyList[2]

        letter1=dummyList[4][0]
        letter2=dummyList[4][1]
        if letter1==dummyList[3]:
            notMinor = letter2
        else:
            notMinor = letter1
        nealeDictLetters[dummyList[0]]=[dummyList[3],notMinor]


    positionList = []
    cntr=0
    for x in range(len(commonList[chrIndex-1])):
        try:
            positionList.append(nealeDict[commonList[chrIndex-1][x]])
        except:
            positionList.append(-1)
            cntr+=1
    print(cntr)

        
    chromosome = 'chr'+str(chrIndex)
        
    cntr = 0
    newCoordSystem = []
    oldCoordSystem = []
    for pos in range(len(positionList)):
        try:
            converted_coordinates = lo.convert_coordinate(chromosome, int(positionList[pos]))
            letter1,letter2 = nealeDictLetters[commonList[chrIndex-1][pos]]


            newName = "chr"+str(chrIndex)+":"+str(converted_coordinates[0][1])+":"+letter1+":"+letter2
            newName2 = "chr"+str(chrIndex)+":"+commonList[chrIndex-1][pos]+":"+letter1+":"+letter2

            newCoordSystem.append(newName)
            oldCoordSystem.append(newName2)
        except:
            cntr+=1

    newCoordList.append(newCoordSystem)
    oldCoordList.append(oldCoordSystem)
```

```python
import pickle

with open('snpPositionsAOU.pkl', 'wb') as f:
    pickle.dump(newCoordList, f)
```
